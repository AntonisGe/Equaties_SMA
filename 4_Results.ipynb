## Data Viz.

```
import pandas as pd
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import datetime
#!pip install yfinance
import yfinance as yf
```

```
df=pd.read_csv('result.csv')
df.head(4)
```

We want to make a 3D graph with those results and change the color of the graph based on the value of result.

So the colors will be:

```
R=0
G=255
B=150
color_list=[]
for i in range(151):
    color_list.append('#%02x%02x%02x' % (R,G,B-i))
R=0
G=255
B=0
for i in range(256):
    color_list.append('#%02x%02x%02x' % (R+i,G,B))
R=255
G=255
B=0
for i in range(256):
    color_list.append('#%02x%02x%02x' % (R,G-i,B))
```

Matching values with colors.

```
value_splits=[(i/len(color_list))*(max(df['result'])-min(df['result']))+min(df['result']) for i in range(len(color_list)+1)]
colors=[]
for i in df['result']:
    for v in range(1,len(value_splits)):
        if value_splits[v-1]<=i and i<=value_splits[v]:
            colors.append(color_list[v-1])
```

The graph.

```
fig = pyplot.figure()

ax = Axes3D(fig)
ax.scatter(df['sma_one'],df['sma_two'],df['result'],c=colors)
ax.grid(False)
ax.set_zticks([])


pyplot.show()
```

```
fig = pyplot.figure()
x = np.linspace(0,50,10000)
y = x
pyplot.plot(x,y,color='black')
ax = pyplot.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
pyplot.ylim(0,120)
pyplot.xlim(0,50)
pyplot.scatter(df['sma_one'],df['sma_two'],c=colors)
pyplot.show()
```

```
best=pd.read_csv('39_112.csv')
best.head(4)
```

So now we want to see if that would be profitable.

Finding first date.

```
first_date=datetime.date(2100,12,12)
for i in best['Date']:
  date_now=datetime.date(int(i.split('-')[0]),int(i.split('-')[1]),int(i.split('-')[2]))
  if first_date>date_now:
    first_date=date_now
print('First date is {}'.format(str(first_date)))
```

We want to compare the strategy to the SP500

```
download=yf.download('SPY',start=str(first_date),end='2020-01-01',progress=False)
sp500=pd.DataFrame()
sp500['Date']=[str(i)[:10] for i in download.index]
sp500['Close']=[i for i in download['Close']]
sp500.tail(4)
```

We want to find two things. One is how many trades would be available each day. And number two compare our profits with SP. From that we can also calculate max dragdown.

```
def convert_string_to_date(string):
  return(datetime.date(int(string.split('-')[0]),int(string.split('-')[1]),int(string.split('-')[2])))

convert_string_to_date('2020-01-01') #Example
```

```
list_of_open_positions_each_day=[]
for i in sp500['Date']:
  number_of_positions_open_that_day=0
  stocks_tested=[]
  for j in range(best.shape[0]):
    if best['Ticker'][j] not in stocks_tested:
      if convert_string_to_date(best['Date'][j])<=convert_string_to_date(i) and convert_string_to_date(best['Last date'][j]) > convert_string_to_date(i):
        stocks_tested.append(best['Ticker'][j])
        number_of_positions_open_that_day += 1
  list_of_open_positions_each_day.append(number_of_positions_open_that_day)
```

```
pyplot.figure(figsize=(20,10))
pyplot.xlim((0,3700))
pyplot.ylim((0,100))
pyplot.plot(list_of_open_positions_each_day[:-2])
ax = pyplot.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax2 = ax.twinx()
pyplot.plot(sp500['Close'],color='orange')
pyplot.ylim((30,350))
ax2.spines['top'].set_visible(False)

pyplot.xticks(ticks=[162,413,665,917,1169,1422,1674,1925,2175,2424,2679,2932,3184,3435,3677],labels=[i for i in range(2006,2021)]);
pyplot.title('Number of Open Positions each day');
```

Yellow line above is SP500 value and the blue is the number of positions we have open each day.

To test our strategy as realistic as possible, we are going to do the following. We start our journey with 116 dollars. Each stock is bought only for 1/100 of our current capital. When the stock is sold our current capital balance changes. If a new stock is to be bought, it will be bought for 1/100 of our current capital.

```
capital=[]
current_capital_balance=sp500['Close'][0]
portofolio=[]
money_spent_on_portofolio=[]
for i in sp500['Date']:
  current_investment=(current_capital_balance+sum(money_spent_on_portofolio))/100
  for j in range(best.shape[0]):
    if i==best['Date'][j]:
      current_capital_balance=current_capital_balance-current_investment
      portofolio.append(best['Ticker'][j])
      money_spent_on_portofolio.append(current_investment)
    if i==best['Date'][j]:
      current_capital_balance = current_capital_balance + best['Profit'][j]/100 + money_spent_on_portofolio[portofolio.index(best['Ticker'][j])]
      del money_spent_on_portofolio[portofolio.index(best['Ticker'][j])]
      portofolio.remove(best['Ticker'][j])
  capital.append(current_capital_balance+sum(money_spent_on_portofolio))

```

```
len(capital)
```

```
pyplot.figure(figsize=(20,10))
pyplot.xlim((0,3700))

pyplot.plot(capital)
ax = pyplot.subplot(111)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
pyplot.plot(sp500['Close'],color='orange')
pyplot.ylim((30,350))


pyplot.xticks(ticks=[162,413,665,917,1169,1422,1674,1925,2175,2424,2679,2932,3184,3435,3677],labels=[i for i in range(2006,2021)]);
pyplot.title('Profits compared to SP 500');
```

Although the strategy is profitable it is not able to beat the S&P, so as expected it is not a good strategy.
